### Showing the frequencies and percentages for columns 'Quantity' and 'Discount'

```python
QD=df.loc[:, ['Quantity', 'Discount']]
```

```python
freqFile = open('freqFile.txt', 'w')

for col in df.select_dtypes(include=["category"]) :
    print(col, "-----------------------------------", 'frequencies',
          df[col].value_counts(), "percentages", df[col].value_counts(normalize=True),
         sep="\n\n", end="\n\n", file=freqFile
         
         )
    


freqFile.close()
```
